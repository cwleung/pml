```python
from distributions.gaussian import MultivariateGaussian

import numpy as np

mvn1 = MultivariateGaussian(np.array([100, 100]), np.array([[100, 0], [0, 100]]))
mvn2 = MultivariateGaussian(np.array([-100, -100]), np.array([[100, 0], [0, 100]]))

mvn3 = mvn1 + mvn2
```

```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots(1, 1, figsize=(12, 12))
ax.scatter(mvn1.sample(100)[:, 0], mvn1.sample(100)[:, 1], label='mvn1')
ax.scatter(mvn2.sample(100)[:, 0], mvn2.sample(100)[:, 1], label='mvn2')
ax.scatter(mvn3.sample(100)[:, 0], mvn3.sample(100)[:, 1], label='mvn3')
ax.legend()
plt.show()
```

```python
# test linear transformation
mvn1 = MultivariateGaussian(np.array([100, 100]), np.array([[100, 0],
                                                            [0, 100]]))
mvn2 = MultivariateGaussian(np.array([-100, -100]), np.array([[100, 0],
                                                              [0, 100]]))

# transformation matrix
A = np.array([[1, 2],
              [2, 1]])

mvn3 = mvn1 @ A

# plot mvn1, mvn2, mvn3, mvn4, mvn5, mvn6 using scatter plot and its samples
import matplotlib.pyplot as plt

fig, ax = plt.subplots(1, 1, figsize=(12, 12))
ax.scatter(mvn1.sample(100)[:, 0], mvn1.sample(100)[:, 1], label='mvn1')
ax.scatter(mvn2.sample(100)[:, 0], mvn2.sample(100)[:, 1], label='mvn2')
ax.scatter(mvn3.sample(100)[:, 0], mvn3.sample(100)[:, 1], label='mvn3')
ax.legend()
plt.show()
```
